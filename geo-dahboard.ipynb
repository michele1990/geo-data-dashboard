```python
# define the stile of the plot using a custom function

def style(p):
    # Custom styling for the plot
    p.title.align = 'center'
    p.title.text_font_size = '20pt'
    p.title.text_font = 'serif'

    p.xaxis.axis_label_text_font_size = '14pt'
    p.xaxis.axis_label_text_font_style = 'bold'
    p.yaxis.axis_label_text_font_size = '14pt'
    p.yaxis.axis_label_text_font_style = 'bold'

    p.xaxis.major_label_text_font_size = '12pt'
    p.yaxis.major_label_text_font_size = '12pt'

    p.legend.location = 'top_left'

    return p
```

```python
# output_notebook()

def plotBar():
    # Data source
    cds = ColumnDataSource(data=dict(
        yrs=years_list,  # Assume years_list is defined elsewhere
        numQuakes=count_list  # Assume count_list is defined elsewhere
    ))

    # Tooltips for interactivity
    TOOLTIPS = [
        ("Year", "@yrs"),
        ("Number of Earthquakes", "@numQuakes")
    ]

    # Create a figure
    barChart = figure(title="Frequency of Earthquakes by Year", 
                      x_axis_label='Years', 
                      y_axis_label='Number of Earthquakes', 
                      width=1000, 
                      height=400, 
                      tooltips=TOOLTIPS,
                      y_range=(0, max(count_list)+100),
                      toolbar_location=None)

    # Add vertical bars
    barChart.vbar(x='yrs', top='numQuakes', width=0.75, color="#009999", legend_label="Year", source=cds)

    # Style the plot
    barChart = style(barChart)

    # Show the plot
    # show(barChart)
    return  barChart
```

```python
# Create the linechart
def plotLine():
    # show the plot imbedded JN
    # output_notebook()

    # load data structure
    cds =  ColumnDataSource(data=dict(
        yrs = years_list,
        numQuakes = count_list
    ))
    #Tooltisp
    TOOLTIPS = [("years","@yrs"), ("Number of earthquackes","@numQuakes") ]

    p = figure(title= "Earthquakes trend by year",
                width=800, 
                height=400, 
                x_axis_label='Years', 
                y_axis_label='Number of Occurencies', 
                x_minor_ticks=2,
                y_range=(0, df_quake_freq["Count"].max() +100),
                toolbar_location=None,
                tooltips=TOOLTIPS)
    p.line(x='yrs',y='numQuakes',color="#009999",line_width=2,legend_label="Yearly Trend", source=cds)
    p.circle(x='yrs',y='numQuakes',color="#009999",size = 8,fill_color = 'white', source=cds)

    # Style the plot
    p = style(p)
    

    # Show the plot
    # show(p)
    return p
```

```python
import math
from math import pi
from bokeh.palettes import Category20c
from bokeh.transform import cumsum

# Create a donut chart
def plotDougnut():
    # output_notebook()
    x = dict(df_load['Type'].value_counts())
    pie_data = pd.Series(x).reset_index(name='value').rename(columns={'index':'type'})


    pie_data['angle'] = pie_data['value']/pie_data['value'].sum() * 2*pi
    pie_data['color'] = Category20c[len(x) if len(x) <= 20 else 20]  # Ensure within range

    p = figure(height=600, 
               title="Types of Earthquakes (1965 - 2026)", 
               toolbar_location=None,  
               tools="hover", 
               tooltips="@type: @value", 
               x_range=(-0.5, 1.0))
    
    p.annular_wedge(x=0, y=1, 
                    inner_radius=0.2, 
                    outer_radius=0.35, 
                    direction="anticlock",
                    start_angle=cumsum('angle', include_zero=True),
                    end_angle=cumsum('angle'),
                    line_color="white",
                    fill_color='color',
                    source=pie_data, 
                    legend_field='type',)
    
    p.axis.axis_label=None
    p.axis.visible=False
    p.grid.grid_line_color = None

    p = style2(p)
    # show(p)
    return p


    

    # # preview the data
    # print(pie_data)


plotDougnut()
```

```python
# Create a magnitde plot
def plotMagnitude():
    # output_notebook()
    magnitude = []

    #get the average magnitude value for each year
    for year in df_quake_freq.Year:
        x = df_load[df_load['Year'] == year]
        data_magnitude = sum(x['Magnitude'])/len(x['Magnitude'])
        magnitude.append(data_magnitude)

    df_quake_freq['Magnitude'] = magnitude

    depth = []
    # get the average depth value for each year
    for year in df_quake_freq.Year:
        x = df_load[df_load['Year'] == year]
        data_depth = sum(x['Depth'])/len(x['Depth'])
        depth.append(data_depth)

    df_quake_freq['Depth'] = depth

    #get the maximum magnitude value for each year
    max_magnitude = list(df_load.groupby('Year')['Magnitude'].max())
    df_quake_freq['Max_Magnitude'] = max_magnitude

    # load the data source
    cds = ColumnDataSource(data=dict(
        yrs = years_list,
        avg_mag = df_quake_freq['Magnitude'].values.tolist(),
        max_mag = df_quake_freq['Max_Magnitude'].values.tolist(),
    ))

    # Tooltips
    TOOLTIPS = [
        ("Year", "@yrs"),
        ("Average Magnitude", "@avg_mag"),
        ("Maximum Magnitude", "@max_mag")
    ]

    mp = figure(title="Average and Maximum Magnitude of Earthquakes by Year",
                width=800,
                height=600,
                x_axis_label='Years',
                y_axis_label='Magnitude',
                x_minor_ticks=2,
                y_range=(5, df_quake_freq["Max_Magnitude"].max() + 1),
                toolbar_location=None,
                tooltips=TOOLTIPS)
    
    # ceate maximum magnitude line
    mp.line(x='yrs', y='max_mag', color="#009999", line_width=2, legend_label="Maximum Magnitude", source=cds)
    mp.circle(x='yrs', y='max_mag', color="#009999", size=8, fill_color='#009999', source=cds)


    # create average magnitude line
    mp.line(x='yrs', y='avg_mag', color="orange", line_width=2, legend_label="Average Magnitude", source=cds)
    mp.circle(x='yrs', y='avg_mag', color="orange", size=8, fill_color='orange', source=cds)

    # Style the plot
    mp = style(mp)

    # Show the plot
    # show(mp)
    return mp


                


    # # preview the data
    # print(df_quake_freq.head())





plotMagnitude()
```

```python
def plotMap():
    # output_notebook()
    # Create a map plot
    lat = df_load['Latitude'].values.tolist()
    lon = df_load['Longitude'].values.tolist()

    lst_lat = []
    lst_lon = []

    i = 0

    # convert the latitude and longitude to mercator coordinates
    for i in range(len(lat)):
        r_major = 6378137.000
        x = r_major * math.radians(lon[i])
        scale = x/lon[i]
        y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 +
                                              lat[i] * math.pi/180.0/2.0)) * scale
        lst_lat.append(y)
        lst_lon.append(x)
        i += 1
    df_load['coordx_x'] = lst_lon
    df_load['coordy_y'] = lst_lat

    lats = df_load['coordy_y'].tolist()
    longs = df_load['coordx_x'].tolist()
    mags = df_load['Magnitude'].tolist()

    # create data source
    cds = ColumnDataSource(data=dict(
        lat = lats,
        lon = longs,
        mag = mags
    ))
    TOOLTIP = [
        ("Magnitude", "@mag")
    ]

    p = figure(title="Earthquakes MAP",
                x_axis_label='Longitude',
                y_axis_label='Latitude',
                width=1000,
                height=600,
                x_range=(-20000000, 6000000),
                y_range=(-10000000, 10000000),
                tooltips=TOOLTIP)

    p.circle(x='lon', y='lat', size=5, color="#009999", alpha=0.5, source=cds, legend_label="Quakes")  

    tile_provider = get_provider(CARTODBPOSITRON)
    p.add_tile(tile_provider)

    p.title.align = 'center'
    p.title.text_font_size = '20px'
    p.title.text_font = 'serif'

    p.legend.location = "bottom_right"
    p.legend.background_fill_color = "white"
    p.legend.background_fill_alpha = 0.5
    # p.legend.label.click_policy="hide"
    # p.legend.label_text_color = "white"

    p.xaxis.visible = False
    p.yaxis.visible = False

    p.axis.axis_label=None
    p.axis.visible=False
    p.grid.grid_line_color = None


    # show(p)
    return p


plotMap()
```

```python
# rename the columns
df_train = df_train.rename(columns={'Magnitude Type': 'Magnitude_Type'})    
df_test_bis = df_test_bis.rename(columns={'Magnitude Type': 'Magnitude_Type'})    
df_test_clean = df_test_clean.rename(columns={'time':'Date','latitude':'Latitude','longitude':'Longitude','mag':'Magnitude', 'depth':'Depth'})
```

```python
# Creating training and testinf Df

df_testing = df_test_clean[['Latitude', 'Longitude', 'Magnitude', 'Depth']]
df_training = df_train[['Latitude', 'Longitude', 'Magnitude', 'Depth']]
df_testing_bis = df_test_bis[['Latitude', 'Longitude', 'Magnitude', 'Depth']]
```

```python
df_testing
```

```python
df_training
```

```python
# remove nulls from the data
df_testing = df_testing.dropna()
df_testing_bis = df_testing_bis.dropna()
df_training = df_training.dropna()
```

```python
# create training data features

X = df_training[['Latitude', 'Longitude']]
y = df_training[['Magnitude', 'Depth']]
```

```python
# create testing data features

X_new = df_testing_bis[['Latitude', 'Longitude']]
y_new = df_testing_bis[['Magnitude', 'Depth']]
```

```python
# import machine learning libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
# use train test split to split the data 20% of the data will be used for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
# create a model
model_reg = RandomForestRegressor(random_state=50)

# Train the model
model_reg.fit(X_train, y_train)

# Predict y_test (magnitude and depth) using x_test features (latitude and longitude)
results = model_reg.predict(X_test)

# check the model accuracy score
score = model_reg.score(X_test, y_test)
```

```python
print("Model Accuracy Score: ", score)
```

```python
# preview the predicted earthquake magnitude and depth
results
```

```python
# improve model accuracy by auto tuning the hyperparameters
parameters = {'n_estimators': [10, 20, 50, 100, 200, 500]}
```

```python
# create GridSearchCV model
grid_obj = GridSearchCV(model_reg, parameters)
# train the model
grid_fit = grid_obj.fit(X_train, y_train)
# get the best model
best_fit = grid_fit.best_estimator_
```

```python
# make the prediction
results = best_fit.predict(X_test)

# check the model accuracy score
score = best_fit.score(X_test, y_test)

print("Model Accuracy Score: ", score)
```

```python
# Predict the earthquake magnitude and depth using the new data
final_results = best_fit.predict(X_new)

# check the model accuracy score
final_score = best_fit.score(X_new, y_new) 
```

```python
print("Model Accuracy Score: ", final_score)
```
